## Initializing Spark Environment

```python
import findspark
findspark.init('/home/shahayush954/spark-3.4.1-bin-hadoop3')
```

```python
from pyspark import SparkContext
```

```python
from pyspark.streaming import StreamingContext
```

```python
sc = SparkContext('local[2]', 'WordCount') # two local threads and name for our context
```

```python
ssc = StreamingContext(sc, 1) # the spark context and the streaming content to be considered in the batches of 1 sec
```

```python
lines = ssc.socketTextStream('localhost', 9999) # the IP and the Port of the Stream to connect. in terminal type nc -lk {portnumber}
                                                # to start giving input at a port
```

```python
words = lines.flatMap(lambda line: line.split(' ')) # take the lines and split by space to get list of words
```

```python
pairs = words.map(lambda word: (word, 1)) # take each word and create a pair of (word, 1)
```

```python
word_counts = pairs.reduceByKey(lambda num1,num2: num1+num2) # for each pair group by firts tupple value and add the second tuple value
```

```python
word_counts.pprint() # pretty print
```

```python
ssc.start()
```
