## Initialize Spark Environment and Imports

```python
import findspark
findspark.init('/home/shahayush954/spark-3.4.1-bin-hadoop3')
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('dateAndTime').getOrCreate()
```

```python
df = spark.read.csv('appl_stock.csv', header=True, inferSchema=True)
```

```python
df.show()
```

## Date Time Manipulation

```python
from pyspark.sql.functions import (dayofmonth, hour, dayofyear, month, year, weekofyear, format_number, date_format)
```

```python
df.select(dayofmonth(df['Date'])).show()
```

```python
df.select(hour(df['Date'])).show()
```

```python
df.select(month(df['Date'])).show()
```

```python
## Average Closing Price per Year Calculation
```

```python
newdf = df.withColumn("Year", year(df['Date']))
```

```python
newdf.show()
```

```python
newdf.groupBy('Year').mean('Close').show()
```
